# Python OOP and System Programming Questions
This notebook contains answers to the following Python-related questions, mainly focusing on exception handling, logging, file handling, and system programming.

### What is the difference between interpreted and compiled languages?
Interpreted languages execute code line-by-line (e.g., Python), while compiled languages convert the entire code to machine code before execution (e.g., C, C++).

### What is exception handling in Python?
It is a mechanism to handle runtime errors using try, except, else, and finally blocks.

### What is the purpose of the finally block in exception handling?
The finally block contains code that will always be executed, regardless of whether an exception occurred.

### What is logging in Python?
Logging records events that happen during execution, useful for debugging and tracking.

### What is the significance of the __del__ method in Python?
It is a destructor method called when an object is about to be destroyed.

### What is the difference between import and from ... import in Python?
`import` imports the whole module; `from ... import` imports specific attributes or functions.

### How can you handle multiple exceptions in Python?
By using multiple except blocks or a tuple of exceptions in a single except block.

### What is the purpose of the with statement when handling files in Python?
`with` ensures proper acquisition and release of resources, such as closing a file after use.

### What is the difference between multithreading and multiprocessing?
Multithreading uses multiple threads within a process; multiprocessing uses multiple processes with separate memory space.

### What are the advantages of using logging in a program?
It helps track errors, provides context, and is useful for debugging and auditing.

### What is memory management in Python?
It's how Python allocates and deallocates memory to objects using reference counting and garbage collection.

### What are the basic steps involved in exception handling in Python?
Use try to write risky code, except to handle exceptions, optionally use else and finally.

### Why is memory management important in Python?
To ensure efficient resource usage, prevent memory leaks, and improve performance.

### What is the role of try and except in exception handling?
`try` contains risky code; `except` handles exceptions that occur.

### How does Python's garbage collection system work?
It uses reference counting and a cyclic garbage collector to free unused memory.

### What is the purpose of the else block in exception handling?
`else` runs if no exception occurs in the try block.

### What are the common logging levels in Python?
DEBUG, INFO, WARNING, ERROR, and CRITICAL.

### What is the difference between os.fork() and multiprocessing in Python?
`os.fork()` is low-level and Unix-specific; `multiprocessing` is a cross-platform module.

### What is the importance of closing a file in Python?
To free system resources and ensure data is properly saved.

### What is the difference between file.read() and file.readline() in Python?
`read()` reads the entire file; `readline()` reads one line at a time.

### What is the logging module in Python used for?
For tracking events, debugging, and recording error messages.

### What is the os module in Python used for in file handling?
To interact with the operating system, including file and directory manipulation.

### What are the challenges associated with memory management in Python?
Circular references, high memory consumption, and fragmentation.

### How do you raise an exception manually in Python?
Using the `raise` keyword with an Exception class.

### Why is it important to use multithreading in certain applications?
For concurrent execution, especially in I/O-bound programs.
